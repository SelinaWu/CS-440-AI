```python
import numpy as np
from copy import deepcopy

#board = np.zeros([7,7])
# board[0][14] = 1

def print_board(board):
    print('  0   1   2   3   4   5   6 ')
    print(' --- --- --- --- --- --- ---')
    for i in range(7):
        temp = []
        temp.append('|')
        for j in range(7):
            if board[i][j] == 0:
                temp.append(' ')
            elif board[i][j] == 1:
                temp.append('@')
            else:
                temp.append('#')
            temp.append('|')
        temp.append(str(i))
        print(' '.join(temp))
        print(' --- --- --- --- --- --- ---')
#         print(' -----------------------------------------------------------')
        
#print_board(board)

#five = 8
#free_four = 7
#dead_four = 6
#free_three = 5
#dead_three = 4
#free_two = 3
#dead_two = 2
#single = 1



def find_winning_block_horiz(board, player):
#     print('#check horizontal')
    free_four_count = 0
    free_three_count = 0
    dead_four_count = 0
    five_count = 0
    dead_three_count = 0
    free_two_count = 0
    dead_two_count = 0


    for i in range(7):
        next_line = 0
        for j in range(3):
            count = 0
            self_count = 0
            consecutive_count = 0
            consecutive_count_max = 0
            first_self = -1
            seven_block = np.zeros(7)
            for k in range(5):
                if board[i][j+k] == player:
                    consecutive_count += 1
                    count+=1
                    self_count+=1
                    seven_block[k+1] = 1
#                     print('fs', first_self, 'k', k)
                    if first_self == -1:
                        first_self = k
#                     print('fs', first_self, 'k', k)
                    
                elif board[i][j+k] == 0:
                    consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
                    consecutive_count = 0
                    count+=1
                    
            consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
            
            if j-1 >= 0:
                if board[i][j-1] != 0 and board[i][j-1] != player:
                    seven_block[0] = -1
            else:
                seven_block[0] = -1
            
            if j+5 <= 6:
                if board[i][j+5] != 0 and board[i][j+5] != player:
                    seven_block[6] = -1
            else:
                seven_block[6] = -1
                    
                    
            if count == 5:
                if self_count == 5:
                    five_count+=1
                    next_line = 1
                    break
                    
                elif self_count == 4:
                    if consecutive_count_max == 4:
                        if seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_four_count+=1
                            next_line  = 1
                            break
                    else:
                        dead_four_count+=1
                        next_line  = 1
                        break
                        
                                
                elif self_count == 3:
                    if consecutive_count_max == 3:
                        if seven_block[first_self] == 0 and seven_block[first_self+4] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                    elif consecutive_count_max == 2:
                        if seven_block[1] == 1 and seven_block[5] == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                        elif seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                    elif consecutive_count_max == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                            
                elif self_count == 2:
                    if consecutive_count_max == 2:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 0:
                            free_two_count+=1
                            next_line = 1
                            break
                        else:
                            dead_two_count+=1
                            next_line = 1
                            break                            
                    else:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 1 and seven_block[first_self+4] == 0:
                            free_two_count+=1
                            next_line = 1
                            break        
                        elif seven_block[first_self] == 0 and seven_block[first_self+4] == 1 and seven_block[first_self+5] == 0:
                            free_two_count+=1
                            next_line = 1
                            break    
            
            if next_line == 1:
                break
       
    #vertical
    for i in range(3):
        next_line = 0
        for j in range(7):
            count = 0
            self_count = 0
            consecutive_count = 0
            consecutive_count_max = 0
            first_self = -1
            seven_block = np.zeros(7)
            for k in range(5):
                if board[i+k][j] == player:
                    consecutive_count += 1
                    count+=1
                    self_count+=1
                    seven_block[k+1] = 1
#                     print('fs', first_self, 'k', k)
                    if first_self == -1:
                        first_self = k
#                     print('fs', first_self, 'k', k)
                    
                elif board[i+k][j] == 0:
                    consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
                    consecutive_count = 0
                    count+=1
                    
            consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
            
            if i-1 >= 0:
                if board[i-1][j] != 0 and board[i-1][j] != player:
                    seven_block[0] = -1
            else:
                seven_block[0] = -1
            
            if i+5 <= 6:
                if board[i+5][j] != 0 and board[i+5][j] != player:
                    seven_block[6] = -1
            else:
                seven_block[6] = -1
                    
                    
            if count == 5:
                if self_count == 5:
                    five_count+=1
                    next_line = 1
                    break
                    
                elif self_count == 4:
                    if consecutive_count_max == 4:
                        if seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_four_count+=1
                            next_line  = 1
                            break
                    else:
                        dead_four_count+=1
                        next_line  = 1
                        break
                        
                                
                elif self_count == 3:
                    if consecutive_count_max == 3:
                        if seven_block[first_self] == 0 and seven_block[first_self+4] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                    elif consecutive_count_max == 2:
                        if seven_block[1] == 1 and seven_block[5] == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                        elif seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                    elif consecutive_count_max == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                            
                elif self_count == 2:
                    if consecutive_count_max == 2:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 0:
                            free_two_count+=1
                            next_line = 1
                            break
                        else:
                            dead_two_count+=1
                            next_line = 1
                            break                            
                    else:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 1 and seven_block[first_self+4] == 0:
                            free_two_count+=1
                            next_line = 1
                            break        
                        elif seven_block[first_self] == 0 and seven_block[first_self+4] == 1 and seven_block[first_self+5] == 0:
                            free_two_count+=1
                            next_line = 1
                            break    
            
            if next_line == 1:
                break
                
    #check right down
    for i in range(3):
        next_line = 0
        for j in range(3):
            count = 0
            self_count = 0
            consecutive_count = 0
            consecutive_count_max = 0
            first_self = -1
            seven_block = np.zeros(7)
            for k in range(5):
                if board[i+k][j+k] == player:
                    consecutive_count += 1
                    count+=1
                    self_count+=1
                    seven_block[k+1] = 1
#                     print('fs', first_self, 'k', k)
                    if first_self == -1:
                        first_self = k
#                     print('fs', first_self, 'k', k)
                    
                elif board[i+k][j+k] == 0:
                    consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
                    consecutive_count = 0
                    count+=1
                    
            consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
            
            if i-1 >= 0 and j-1 >= 0:
                if board[i-1][j-1] != 0 and board[i-1][j-1] != player:
                    seven_block[0] = -1
            else:
                seven_block[0] = -1
            
            if i+5 <= 6 and j+5 <= 6:
                if board[i+5][j+5] != 0 and board[i+5][j+5] != player:
                    seven_block[6] = -1
            else:
                seven_block[6] = -1
                    
                    
            if count == 5:
                if self_count == 5:
                    five_count+=1
                    next_line = 1
                    break
                    
                elif self_count == 4:
                    if consecutive_count_max == 4:
                        if seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_four_count+=1
                            next_line  = 1
                            break
                    else:
                        dead_four_count+=1
                        next_line  = 1
                        break
                        
                                
                elif self_count == 3:
                    if consecutive_count_max == 3:
                        if seven_block[first_self] == 0 and seven_block[first_self+4] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                    elif consecutive_count_max == 2:
                        if seven_block[1] == 1 and seven_block[5] == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                        elif seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                    elif consecutive_count_max == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                            
                elif self_count == 2:
                    if consecutive_count_max == 2:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 0:
                            free_two_count+=1
                            next_line = 1
                            break
                        else:
                            dead_two_count+=1
                            next_line = 1
                            break                            
                    else:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 1 and seven_block[first_self+4] == 0:
                            free_two_count+=1
                            next_line = 1
                            break        
                        elif seven_block[first_self] == 0 and seven_block[first_self+4] == 1 and seven_block[first_self+5] == 0:
                            free_two_count+=1
                            next_line = 1
                            break    
            
            if next_line == 1:
                break
                
                
        #check right down
    for i in range(6,3,-1):
        next_line = 0
        for j in range(3):
            count = 0
            self_count = 0
            consecutive_count = 0
            consecutive_count_max = 0
            first_self = -1
            seven_block = np.zeros(7)
            for k in range(5):
                if board[i-k][j+k] == player:
                    consecutive_count += 1
                    count+=1
                    self_count+=1
                    seven_block[k+1] = 1
#                     print('fs', first_self, 'k', k)
                    if first_self == -1:
                        first_self = k
#                     print('fs', first_self, 'k', k)
                    
                elif board[i-k][j+k] == 0:
                    consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
                    consecutive_count = 0
                    count+=1
                    
            consecutive_count_max = consecutive_count if consecutive_count > consecutive_count_max else consecutive_count_max
            
            if i+1 <= 6 and j-1 >= 0:
                if board[i+1][j-1] != 0 and board[i+1][j-1] != player:
                    seven_block[0] = -1
            else:
                seven_block[0] = -1
            
            if i-5 >= 0 and j+5 <= 6:
                if board[i-5][j+5] != 0 and board[i-5][j+5] != player:
                    seven_block[6] = -1
            else:
                seven_block[6] = -1
                    
                    
            if count == 5:
                if self_count == 5:
                    five_count+=1
                    next_line = 1
                    break
                    
                elif self_count == 4:
                    if consecutive_count_max == 4:
                        if seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_four_count+=1
                            next_line  = 1
                            break
                    else:
                        dead_four_count+=1
                        next_line  = 1
                        break
                        
                                
                elif self_count == 3:
                    if consecutive_count_max == 3:
                        if seven_block[first_self] == 0 and seven_block[first_self+4] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                    elif consecutive_count_max == 2:
                        if seven_block[1] == 1 and seven_block[5] == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                        elif seven_block[first_self] == 0 and seven_block[first_self+5] == 0:
                            free_three_count+=1
                            next_line = 1
                            break
                        else:
                            dead_three_count+=1
                            next_line = 1
                            break
                                    
                    elif consecutive_count_max == 1:
                            dead_three_count+=1
                            next_line = 1
                            break
                            
                elif self_count == 2:
                    if consecutive_count_max == 2:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 0:
                            free_two_count+=1
                            next_line = 1
                            break
                        else:
                            dead_two_count+=1
                            next_line = 1
                            break                            
                    else:
                        if seven_block[first_self] == 0 and seven_block[first_self+3] == 1 and seven_block[first_self+4] == 0:
                            free_two_count+=1
                            next_line = 1
                            break        
                        elif seven_block[first_self] == 0 and seven_block[first_self+4] == 1 and seven_block[first_self+5] == 0:
                            free_two_count+=1
                            next_line = 1
                            break    
            
            if next_line == 1:
                break
                
    res = np.zeros(2)
    
    if five_count > 0:
        res[0] += 1000000000
    elif free_four_count > 0:
        res[0] += 100000000
    elif dead_four_count >= 2:
        res[0] += 10000000
    elif dead_four_count == 1 and free_three_count > 0:
        res[0] += 1000000
    elif free_three_count >= 2:
        res[0] += 100000
    elif free_three_count == 1:
        res[0] += 10000
        
    res[1] += dead_three_count * 10 
    res[1] += free_two_count * 5
    res[1] += dead_two_count * 2
        
    print
        
    return res
                
    
    
def evaluation_wrapper(board, max_agent):
    return find_winning_block_horiz(board,max_agent) - find_winning_block_horiz(board, max_agent%2+1)

def less_than(left, right):
    temp = True
    if left[0] > right[0]:
        temp = False
    elif left[0] == right[0] and left[1] > right[1]:
        temp = False
    return temp

def AlphaBeta(board, max_val):
    # this function would return the index = i * 15 + j
    
    # determine what is the max_val and min_val
    # max is the agent, min is the opponent
    # CASE1: max_val = 1, min_val = 2
    # CASE2: max_val = 2, min_val = 1
    if max_val == 1:
        min_val = 2
    elif max_val == 2:
        min_val = 1
    else:
        print('AlphaBeta::max_val input is not correct!')
        return -1
    
    # First is max move, Second is the min move, Thrid is the max move
    expand_count = 0
    
    # Searching through all the first move
    move1_idx = -1
    move1_eval = [-np.inf,-np.inf]
    for idx1 in range(49):
        if board[idx1//7][idx1%7] == 0:
            expand_count += 1
#             print(expand_count)
            # set the current move and find its evaluation
            board[idx1//7][idx1%7] = max_val
            
            # Searching through all the second move
            move2_idx = -1
            move2_eval = [np.inf, np.inf]
            for idx2 in range(49):
                if not less_than(move1_eval , move2_eval):
                    break
                if board[idx2//7][idx2%7] == 0:
                    expand_count += 1
                    
                    # set the current move and find its evaluation
                    board[idx2//7][idx2%7] = min_val
    
                    #Searching through all the thrid move
                    move3_idx = -1
                    move3_eval = [-np.inf,-np.inf]
                    for idx3 in range(49):
                        if less_than(move2_eval, move3_eval):
                            break
                        if board[idx3//7][idx3%7] == 0:
                            expand_count += 1
                            
                            # set the current move and find its evaluation
                            board[idx3//7][idx3%7] = max_val
                            temp = evaluation_wrapper(board, max_val)
                            #temp = 0
                            
                            # if the current eval is larger
                            if not less_than(temp , move3_eval):
                                move3_eval = temp
                                move3_idx = idx3
                            
                            # reset the current move
                            board[idx3//7][idx3%7] = 0
                    
                    # if the current eval is smaller
                    if less_than(move3_eval , move2_eval):
                        move2_eval = move3_eval
                        move2_idx = idx2
                            
                    # reset the current move
                    board[idx2//7][idx2%7] = 0
            
            # if the current eval is larger
            if not less_than(move2_eval , move1_eval):
                move1_eval = move2_eval
                move1_idx = idx1
                            
            # reset the current move
            board[idx1//7][idx1%7] = 0
#     print(expand_count)
#     print(move1_eval)
    print(move1_idx)
#     print(move2_eval)
#     print(move2_idx)
#     print(move3_eval)
#     print(move3_idx)
    return move1_idx

def check_four(board, target):
    r = target//7
    c = target%7
    if board[r][c] == 0:
        return -1, 0, 0
    next_ = -1
    #check horizontal
    i = c
    count = 0
    is_open = 2
    while i >= 0:#left
        if board[r][i] == board[r][c]:
            count+=1
            i-=1
#             print("left")
        elif board[r][i] != 0:
            is_open-=1
            break
        else:
            next_ = r*7+i
            break            
    i = c
    count-=1
    while i < 7:#right
        if board[r][i] == board[r][c]:
            count+=1
            i+=1
#             print("right")
        elif board[r][i] != 0:
            is_open-=1
            break
        else:
            next_ = r*7+i
            break
    if count == 4 and is_open > 0:
        return 1,next_//7, next_%7
        
    #check vertical
    i = r
    next_ = -1
    count = 0
    is_open = 2
    while i >= 0:#up
        if board[i][c] == board[r][c]:
            count+=1
            i-=1
#             print("up")
        elif board[i][c] != 0:
            is_open-=1
#             print("ASASASA")
            break
        else:
            next_ = i*7+c
            break
        
    i = r
    count-=1
    while i < 7:#down
        if board[i][c] == board[r][c]:
            count+=1
            i+=1
#             print("down")
        elif board[i][c] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+c
            break
#     print(is_open)
    if count == 4 and is_open > 0:
        return 1,next_//7, next_%7
        
    #check left up and right down
    i = r 
    j = c
    next_ = -1
    count = 0
    is_open = 2
    while i >= 0 and j>=0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j-=1
#             print("left up")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
            break

    i = r
    j = c
    count-=1
    while i < 7 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j+=1
#             print("down right")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
#             print(next_)
            break
    if count == 4 and is_open > 0:
        return 1,next_//7, next_%7
        
    #check left down and right up
    i = r 
    j = c
    next_ = -1
    count = 0
    is_open = 2
    while i < 7 and j >= 0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j-=1
#             print("left down")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
            break

    i = r
    j = c
    count-=1
    while i >= 0 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j+=1
#             print("right up")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
            break
    if count == 4 and is_open > 0:
        return 1,next_//7, next_%7
        
    return 0,0,0
    
def check_three(board, target):
    r = target//7
    c = target%7
    if board[r][c] == 0:
        return -1, 0,0
    next_ = -1
    #check horizontal
    i = c
    count = 0
    is_open = 2
    while i >= 0:#left
        if board[r][i] == board[r][c]:
            count+=1
            i-=1
#             print("left")
        elif board[r][i] != 0:
            is_open-=1
            break
        else:
            next_ = r*7+i
            break
            
    if i == -1:
        is_open-=1
        
    i = c
    count-=1
    while i < 7:#right
        if board[r][i] == board[r][c]:
            count+=1
            i+=1
#             print("right")
        elif board[r][i] != 0:
            is_open-=1
            break
        else:
            if next_ == -1:
                next_ = r*7+i
            break
            
    if i == 7:
        is_open -= 1
        
    if count == 3 and is_open == 2:
        return 1,next_//7, next_%7
        
    #check vertical
    i = r
    count = 0
    next_ = -1
    is_open = 2
    while i >= 0:#up
        if board[i][c] == board[r][c]:
            count+=1
            i-=1
#             print("up")
        elif board[i][c] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+c
            break
        
    if i == -1:
        is_open -= 1
    
    i = r
    count-=1
    while i < 7:#down
        if board[i][c] == board[r][c]:
            count+=1
            i+=1
#             print("down")
        elif board[i][c] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+c
            break
            
    if i == 7:
        is_open -= 1
            
    if count == 3 and is_open == 2:
        return 1,next_//7, next_%7
        
    #check left up and right down
    i = r 
    j = c
    next_ = -1
    count = 0
    is_open = 2
    while i >= 0 and j>=0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j-=1
#             print("left up")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
            break
    
    if i == -1:
        is_open -= 1
    
    i = r
    j = c
    count-=1
    while i < 7 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j+=1
#             print("down right")
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            if next_ == -1:
                next_ = i*7+j
#             print(next_)
            break
    
    if i == 7:
        is_open -= 1
    
    if count == 3 and is_open == 2:
        return 1,next_//7, next_%7
        
    #check left down and right up
    i = r 
    j = c
    next_ = -1
    count = 0
    is_open = 2
    while i < 7 and j >= 0:#left up
#         print("left down",i,j)
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j-=1
            
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            next_ = i*7+j
            break
            
    if i == 7 or j == -1:
        is_open -= 1

    i = r
    j = c
    count-=1
    while i >= 0 and j < 7:#right down
#         print("right up", i,j)
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j+=1
            
        elif board[i][j] != 0:
            is_open-=1
            break
        else:
            if next_ == -1:
                next_ = i*7+j
            break
     
    if i == -1 or j == 7:
        is_open -= 1
    
    if count == 3 and is_open == 2:
        return 1,next_//7, next_%7
        
    return 0,0,0
    
def check_five(board, target):
    r = target//7
    c = target%7
    if board[r][c] == 0:
        return -1 
    
    #check horizontal
    i = c
    count = 0

    while i >= 0:#left
        if board[r][i] == board[r][c]:
            count+=1
            i-=1
#             print("left")
        else:
            break 
            
    i = c
    count-=1
    
    while i < 7:#right
        if board[r][i] == board[r][c]:
            count+=1
            i+=1
#             print("right")
        else:
            break
    if count >= 5:
        return 1
        
    #check vertical
    i = r
    count = 0

    while i >= 0:#up
        if board[i][c] == board[r][c]:
            count+=1
            i-=1
#             print("up")
        else:
            break
        
    i = r
    count-=1
    while i < 7:#down
        if board[i][c] == board[r][c]:
            count+=1
            i+=1
#             print("down")
        else:
            break

    if count >= 5:
        return 1
        
    #check left up and right down
    i = r 
    j = c
    count = 0
    
    while i >= 0 and j>=0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j-=1
#             print("left up")
        else:
            break

    i = r
    j = c
    count-=1
    while i < 7 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j+=1
#             print("down right")
        else:
            break
    if count >= 5:
        return 1
        
    #check left down and right up
    i = r 
    j = c
    count = 0

    while i < 7 and j >= 0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j-=1
#             print("left down")
        else:
            break

    i = r
    j = c
    count-=1
    while i >= 0 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j+=1
#             print("right up")
        else:
            break
    if count >= 5:
        return 1
        
    return 0

#from copy import deepcopy
def find_winning_block(board, player):
    winning_set = [-1,-1,-1]
    
#     print('#check horizontal')
    for i in range(7):
        for j in range(3):
            count = 0
            self_count = 0
            first_self = -1
            for k in range(5):
                if board[i][j+k] == player:
                    count+=1
                    self_count+=1
#                     print('fs', first_self, 'k', k)
                    if first_self == -1:
                        first_self = k
#                     print('fs', first_self, 'k', k)
                    
                elif board[i][j+k] == 0:
                    count+=1
            if count == 5:
                if first_self == 0:
                    while(board[i][j+first_self] == player):
                        first_self+=1
                    next_r = i
                    next_c = j+first_self
                elif first_self != -1:
                    next_r = i
                    next_c = j+first_self-1
                else:
                    next_r = i
                    next_c = j

                temp = []
                temp.append(self_count)
                temp.append(next_r)
                temp.append(next_c)
#                 print(temp)
                if self_count == winning_set[0]:
                    if next_c < winning_set[2]:
                        winning_set = temp
#                         print(winning_set, i , j ,first_self)
                    elif next_c == winning_set[2]:
                        if next_r > winning_set[1]:
                            winning_set = temp
#                             print(winning_set, i , j ,first_self)
                            
                elif self_count > winning_set[0]:
                    winning_set = temp
#                     print(winning_set, i , j ,first_self)
        
#     print('#check vertical')
    for i in range(3):
        for j in range(7):
            count = 0
            self_count = 0
            first_self = -1
            for k in range(5):
                if board[i+k][j] == player:
                    count+=1
                    self_count+=1
                    first_self = k
                    
                elif board[i+k][j] == 0:
                    count+=1
            if count == 5:
                if first_self == 4:
                    while(board[i+first_self][j] == player):
                        first_self-=1
                    next_r = i+first_self
                    next_c = j
                elif first_self != -1:
                    next_r = i+first_self+1
                    next_c = j
                else:
                    next_r = i
                    next_r = j
                    
                temp = []
                temp.append(self_count)
                temp.append(next_r)
                temp.append(next_c)
                
                if self_count == winning_set[0]:
                    if next_c < winning_set[2]:
                        winning_set = temp
#                         print(winning_set, i , j ,first_self)
                    elif next_c == winning_set[2]:
                        if next_r > winning_set[1]:
                            winning_set = temp
#                             print(winning_set, i , j ,first_self)
                            
                elif self_count > winning_set[0]:
                    winning_set = temp
#                     print(winning_set, i , j ,first_self)
                    
#     print('#check right down')
    for i in range(3):
        for j in range(3):
            count = 0
            self_count = 0
            first_self = -1
            for k in range(5):
                if board[i+k][j+k] == player:
                    count+=1
                    self_count+=1
                    if first_self == -1:
                        first_self = k
                    
                elif board[i+k][j+k] == 0:
                    count+=1
            if count == 5:
                if first_self == 0:
                    while(board[i+first_self][j+first_self] == player):
                        first_self+=1
                    next_r = i+first_self
                    next_c = j+first_self
                elif first_self != -1:
                    next_r = i+first_self-1
                    next_c = j+first_self-1
                else:
                    next_r = i
                    next_r = j
                    
                temp = []
                temp.append(self_count)
                temp.append(next_r)
                temp.append(next_c)
                
                if self_count == winning_set[0]:
                    if next_c < winning_set[2]:
                        winning_set = temp
#                         print(winning_set, i , j ,first_self)
                    elif next_c == winning_set[2]:
                        if next_r > winning_set[1]:
                            winning_set = temp
#                             print(winning_set, i , j ,first_self)
                            
                elif self_count > winning_set[0]:
                    winning_set = temp
#                     print(winning_set, i , j ,first_self)
                    
    #check left up
#     print('#check left up')
    for i in range(3):
        for j in range(3):
            count = 0
            self_count = 0
            first_self = -1
            for k in range(5):
                if board[6-i-k][j+k] == player:
                    count+=1
                    self_count+=1
                    if first_self == -1:
                        first_self = k
                    
                elif board[6-i-k][j+k] == 0:
                    count+=1
            
            if count == 5:
                if first_self == 0:
                    while(board[6-i-first_self][j+first_self] == player):
                        first_self+=1
                    next_r = 6-i-first_self
                    next_c = j+first_self
                elif first_self != -1:
                    next_r = 6-i-first_self+1
                    next_c = j+first_self-1
                else:
                    next_r = 6 - i
                    next_c = j
                    
                temp = []
                temp.append(self_count)
                temp.append(next_r)
                temp.append(next_c)
                
                if self_count == winning_set[0]:
                    if next_c < winning_set[2]:
                        winning_set = temp
#                         print(winning_set, i , j ,first_self)
                    elif next_c == winning_set[2]:
                        if next_r > winning_set[1]:
                            winning_set = temp
#                             print(winning_set, i , j ,first_self)
                            
                elif self_count > winning_set[0]:
                    winning_set = temp
#                     print(winning_set, i , j ,first_self)
                    
    return winning_set[1], winning_set[2]
                    
def reflex_agent_wrapper(board, agent):
    next_step = [-1, 7]
#     print('***************************step 1*******************************')
    for j in range(49):
        r = j//7
        c = j%7
        if board[r][c] == agent:
            temp = check_four(board, r*7+c)
            if temp[0] == 1:
                if next_step[1] > temp[2]:
                    next_step[0] = temp[1]
                    next_step[1] = temp[2]
                    print(next_step)
                elif next_step[1] == temp[2]:
                    if next_step[0] < temp[1]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
#     print('***************************step 2*******************************')
    if next_step[0] == -1:
        for j in range(49):
            r = j//7
            c = j%7
            if board[r][c] != agent and board[r][c] != 0:
                temp = check_four(board, r*7+c)
                if temp[0] == 1:
                    if next_step[1] > temp[2]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
                    elif next_step[1] == temp[2]:
                        if next_step[0] < temp[1]:
                            next_step[0] = temp[1]
                            next_step[1] = temp[2]
                            print(next_step)
#     print('***************************step 3*******************************')
    if next_step[0] == -1:
        for j in range(49):
            r = j//7
            c = j%7
            if board[r][c] != agent and board[r][c] != 0:
                temp = check_three(board, r*7+c)
                if temp[0] == 1:

                    if next_step[1] > temp[2]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
                    elif next_step[1] == temp[2]:
                        if next_step[0] < temp[1]:
                            next_step[0] = temp[1]
                            next_step[1] = temp[2]  
                            print(next_step)
#     print('***************************step 4*******************************')
    if next_step[0] == -1:
        r,c = find_winning_block(board, agent)
        if r != -1 and c != -1:
            next_step[0] = r
            next_step[1] = c
            print(next_step)
#     print('***************************step 5*******************************')  
    if next_step[0] == -1:
        for c in range(7):
            for r in range(7):
                if board[r][c] == 0:
                    next_step[0] = r
                    next_step[1] = c
#                    print(next_step)
                    break
            if next_step[0] != -1:
                break
#     print('***************************step 6*******************************')
    board[next_step[0]][next_step[1]] = agent
#    print(agent, 'next_step is', next_step)
    return next_step[0]*7+next_step[1]

```

```python
from tkinter import *
import numpy as np

board = np.zeros([7,7])
winner = [0]

window = Tk()
window.title('Gomoku')

c_dim = 349
c_space = (c_dim+1)//7
canvas = Canvas(window, height=c_dim, width=c_dim)
for i in range(6):
    canvas.create_line(c_space*(i+1)-1, 0, c_space*(i+1)-1, c_dim)
    canvas.create_line(0, c_space*(i+1)-1, c_dim, c_space*(i+1)-1)
canvas.pack()

status = Label(window, text="Waiting for the palyer's move...", bd=1, relief=SUNKEN, anchor=W)
status.pack(side=BOTTOM, fill=X)

def leftClick(event):
    #print(event.x, event.y)
    x = event.x//c_space
    y = event.y//c_space
    #print(x, y)
    
    if board[y][x] == 0 and winner[0] == 0:
        canvas.create_oval(c_space*(x+0.2), c_space*(y+0.2), c_space*(x+0.8), c_space*(y+0.8), fill="red")
        board[y][x] = 1
        if check_five(board, (y*7 + x)) > 0:
            winner[0] = 1
        if winner[0] == 1:
            status['text'] = "Player is the winner!"
        if winner[0] == 0:
            status['text'] = "Waiting for the AI's move..."
            AI_move = AlphaBeta(board, 2)
            AI_x = AI_move%7
            AI_y = AI_move//7
            board[AI_y][AI_x] = 2
            canvas.create_oval(c_space*(AI_x+0.2), c_space*(AI_y+0.2), c_space*(AI_x+0.8), c_space*(AI_y+0.8), fill="blue")
            if check_five(board, AI_move) > 0:
                winner[0] = 2
            status['text'] = "Waiting for the palyer's move..."
            if winner[0] == 2:
                status['text'] = "AI is the winner!"
    elif board[y][x] != 0 and winner[0] == 0:
        status['text'] = "Invalid input!"

def rightClick(event):
    canvas.delete(ALL)
    for i in range(6):
        canvas.create_line(c_space*(i+1)-1, 0, c_space*(i+1)-1, c_dim)
        canvas.create_line(0, c_space*(i+1)-1, c_dim, c_space*(i+1)-1)
    canvas.pack()
    for i in range(7):
        for j in range(7):
            board[i][j] = 0
    winner[0] = 0
    status['text'] = "Waiting for the palyer's move..."
                
canvas.bind("<Button-1>", leftClick)
canvas.bind("<Button-2>", rightClick)

window.mainloop()
```
